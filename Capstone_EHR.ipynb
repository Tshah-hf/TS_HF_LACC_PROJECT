#  Add & print the Month name based on Payment Date

```python
df['Month']= pd.DatetimeIndex(df['Payment Date']).month_name(locale='English')
df.head(10)
```

# Grouping by  'Payment Year' to calculate 'Total payments'

```python
des=df.groupby(['Payment Year'])['Payment Amount'].sum()
des
```

```python
# matplotlib line chart, to understand which was highest payment year
   
Year = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
Payment_Amount = [171142067,242629406,186566291,128664026,85270348,109328261,32679566,32341861,20910004,18810504,15637169]
  
plt.plot(Year, Payment_Amount, color='red',marker='o')
plt.title('Total Payments By Year', fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Payment_Amount',fontsize=15)
plt.show()
```

# grouping by  'Month' to see which month was total payments at highest

# grouping by  'Year' and 'practioner name' to see which year was hightest distinct practitioner participation

```python
# grouping by  'Year' and 'practioner name' to see which year was hightest distinct practitioner participation
df1 = df.groupby('Payment Year').agg(
    {'Practitioner_Count': np.sum, 'Practitioner_Name': pd.Series.nunique})
df1
```

```python
#graph matplotlib
plt.style.use('ggplot')
x =['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
Total = [3373,5385,6125,6619,5921,7373,3624,3752,2458,2213,1843]
x_pos =[i for i,_ in enumerate(x)]

plt.bar(x_pos, Total, color='Blue')
plt.xlabel("Year")
plt.ylabel("Distinct_Practitioner")
plt.title('Distinct Practitioner Participation By Year')
plt.xticks(x_pos, x)
plt.figure(figsize=(20,40))
# plt.bar['figure.figsize']= [12,8]
plt.show()
```

# Total payment and distict count of practitioner name and distinct count of county by Year

```python
df2 = df.groupby('Payment Year').agg(
    {'Payment Amount': np.sum, 'Practitioner_Name': pd.Series.nunique, 'County': pd.Series.nunique })
df2
```

```python
# grouping by  'County' to see which top 10 county has total payments at highest
Con=df.groupby(['County'])['Payment Amount'].sum().sort_values(ascending=False).reset_index().head(10)
Con
```

```python
y=["NEW YORK","BRONX","KINGS","QUEENS","WESTCHESTER","MONROE","NASSAU","SUFFOLK","ONONDAGA","ERIE"]
 
# getting values against each value of y
x=[214831664,147098643,143446865,93764468,53467521,43856726,42525918,39792163,33633333,31686808]
plt.barh(y, x)
 
# setting label of y-axis
plt.ylabel("Top 10 County")
 
# setting label of x-axis
plt.xlabel("Total Payment")
plt.title("Top 10 County By Total Incentive")
plt.show()
```

```python
PT = df.loc[(df['Provider Type']=='ACUTE CARE HOSPITALS')]
```

```python
PT
```

```python
# grouping by  'Provider type' to see which highest paid hospital
Hos=PT.groupby(['Eligible Hospital Name'])['Payment Amount'].sum().sort_values(ascending=False).reset_index().head(10)
Hos
```

```python
df = pd.DataFrame(Hos,columns=['Eligible Hospital Name','Payment Amount'])
  
plt.plot(df['Eligible Hospital Name'], df['Payment Amount'], color='red', marker='o')
plt.title('Payment By Hospitals', fontsize=14)
plt.xlabel('Eligible Hospital Name', fontsize=14)
plt.xticks(rotation=90)
plt.ylabel('Payment Amount', fontsize=14)
plt.grid(True)
plt.show()
```
