```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
df = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")
```

```python
drop_cols = ['School DBN', 'Percent Scoring Below 65', 'Percent Scoring 65 or Above', 'Percent Scoring 80 or Above', 'Number Scoring CR', 'Percent Scoring CR', 'Mean Score', 'Number Scoring Below 65', 'Number Scoring 65 or Above']
df.drop(drop_cols, inplace = True, axis = 1)
```

```python
df.head()
```

```python
df2 = df.loc[(df['School Name']=='Orchard Collegiate Academy')& (df['Regents Exam']=='Common Core Algebra')]
df2
```

```python
df2.info()
```

```python
df2['Number Scoring 80 or Above']=pd.to_numeric(df2['Number Scoring 80 or Above'],errors='coerce')
df2
```

```python
df2.info()
```

```python
df2['Number Scoring 80 or Above'].fillna(0,inplace=True)
df2
```

```python
#Descriptive statistics
df2.describe()
```

```python
#sort and group
df2.sort_values(['Year','Total Tested'], ascending=[True,False])
```

```python
df3 = df2.groupby(['Year'])['Total Tested'].sum().sort_values(ascending=True)
df3
```

```python
df4=df2.groupby(['Year'])['Number Scoring 80 or Above'].sum().sort_values(ascending=True)
df4
```

```python
#graph matplotlib
plt.style.use('ggplot')
x =['2015','2016','2017']
Total = [60,250,265]
x_pos =[i for i,_ in enumerate(x)]

plt.bar(x_pos, Total, color='pink')
plt.xlabel("Year")
plt.ylabel("Total Tested")
plt.title('Orchard Collegiate Academy-Common Core Algebra- Total Tested By Year')
plt.xticks(x_pos, x)
plt.show()
```
